## Step 0 - Read data and Analyze its quality 

### Imbalanced Data? 

Only %14.38 of the train data is labeled as **1**, while the remaining ones are examples for **0**'s. To address this issue, I will follow the method suggested by *Wei-Chao et al, 2017* in their work **Clustering-based undersampling in class-imbalanced data**. 

I will undersample the majority class by trainin a KMeans cluster on it. Specifically, the number of clusters in the majority class is set to be equal to the number of data points in the minority class. Then the cluster centers will be used to represent the majority class in the classification stage. 

This method significantly increases the classfication abilities of all the models considered in this study. For instance, AUC score  of the Random Forest Classifier increased from 0.79 to 0.88. 

It's important to note that this down-sampling technique is conducted solely on the training set, and both the validation and test sets are left untouched by this process.

### Data Quality

Following categorical variables had the explained data quality issues and are handled during data cleaning process. 
- **x3** : Both short-hand and full version of the days of the week were present. All converted to their full version, e.g. 'Wed' -> 'Wednesday', 
- **x7** : (float) percentages were displayed as string due to '\%' sign. Converted to float type (in terms of percentages), e.g. 
    '0.0062\%' ==> 0.0062, 
- **x19**: (float) dollar amounts were displayed as string due to dollar sign. Converted to float type (in terms of percentages), e.g. '&#36;-908.650758' ==> -908.650758, 
- **x33** and **x99** had no variation in data, hence omitted, 

The same cleaning is applied to test data as well. 

During visual exploration, I noticed that the columns **x59**, **x79**, and **x98** were binary, so converted them to object types later in the analysis -- both in dev, val, and test sets of course.

## Step 4 - Compare the modeling approaches:

### Logistic Regression vs Random Forest  

Even though I considered multiple models, I end up choosing a Random Forest Classifier predictor. 

Random Forest Classifier and Logistic Regression Classifier are two popular algorithms used in the field of machine learning for classification tasks. While both algorithms have their strengths and weaknesses, there are some significant differences between them.

Here are some of the key differences between Random Forest Classifier and Logistic Regression Classifier:

- **Algorithm:** Random Forest is an ensemble learning algorithm that combines multiple decision trees to make a final prediction. In contrast, logistic regression is a linear model that models the relationship between the dependent variable and the independent variables by estimating probabilities.

- **Interpretability:** Logistic Regression is more interpretable than Random Forest. It provides a straightforward understanding of the coefficients for each independent variable and how they impact the dependent variable. On the other hand, Random Forest is a black box algorithm, and it can be challenging to interpret how individual trees impact the final prediction.

- **Handling Missing Data:** Random Forest can handle missing data and imputes it by creating surrogate splits. Logistic Regression requires complete data, and missing values must be imputed before fitting the model.

- **Handling Non-linear Relationships:** Random Forest can handle non-linear relationships between the dependent and independent variables, while Logistic Regression assumes a linear relationship between them.

- **Over-fitting:** Random Forest can suffer from over-fitting if the number of trees is too high, while logistic regression is less prone to over-fitting. 

- **Feature Importance:** Random Forest can provide feature importance, which can be useful for feature selection. Logistic Regression does not provide feature importance, but the coefficients can give an idea of which independent variables are more important in the model.

In summary, Random Forest and Logistic Regression are two different algorithms with different strengths and weaknesses. Logistic Regression is more interpretable and better suited for linear relationships, while Random Forest is better suited for handling non-linear relationships and providing feature importance. Choosing the right algorithm depends on the problem at hand and the nature of the data.

## Step 1 - Clean and prepare data

### Data Cleaning

```python
def clean_days(X):
    days_conversion = {
        "Mon": "Monday",
        "Tue": "Tuesday",
        "Wed": "Wednesday",
        "Thur": "Thursday",
        "Fri": "Friday",
        "Sat": "Saturday",
        "Sun": "Sunday",
    }
    # Map the short-hand days to their full version, e.g. 'Wed' -> 'Wednesday'
    X["x3"] = X["x3"].map(days_conversion).fillna(X["x3"])
    return X


def clean_percentage_signs(X):
    # convert (string) percentages to floats, e.g. '0.0062%' - > 0.0062
    X["x7"] = X["x7"].str.replace("%", "").astype(float)
    return X


def clean_dollar_signs(X):
    # convert (string) dollars to floats, e.g. '$-908.650758' - >-908.650758
    X["x19"] = X["x19"].str.replace("$", "").astype(float)
    return X


def clean_useless_columns(X):
    # Drop the columns that do not carry information/variation
    X.drop(["x39", "x99"], axis=1, inplace=True)
    return X


def clean_data(X):
    X = clean_days(X)
    X = clean_percentage_signs(X)
    X = clean_dollar_signs(X)
    X = clean_useless_columns(X)
    return X
```

```python
def feature_scaling_boxcox(df):
    # transform these columns to look more Gaussian
    df["x42"] = df["x42"].apply(lambda x: boxcox(x, 0.5))
    df["x45"] = df["x45"].apply(lambda x: boxcox(x, 0.5))
    df["x55"] = df["x55"].apply(lambda x: boxcox(x, 0.3))
    df["x96"] = df["x96"].apply(lambda x: boxcox(x, -1.5))
    df["x12"] = df["x12"].apply(lambda x: boxcox(x, 0.3))
    df["x14"] = df["x14"].apply(lambda x: boxcox(x, 0.5))
    df["x18"] = df["x18"].apply(lambda x: boxcox(x, -1.5))
    df["x21"] = df["x21"].apply(lambda x: boxcox(x, 0.5))
    df["x40"] = df["x40"].apply(lambda x: boxcox(x, -1.5))
    df["x61"] = df["x61"].apply(lambda x: boxcox(x, 0))
    return df
```

```python
# Clean both train and test data
train_data = clean_data(train_data)
test_data = clean_data(test_data)

# Transform non-Gaussian columns
train_data = feature_scaling_boxcox(train_data)
test_data = feature_scaling_boxcox(test_data)
train_data.head()
```

## Transformers

```python
# PREPROCESS
num_cols = [col for col in X_train.columns if X_train[col].dtype in ["int64", "float"]]
cat_cols = [col for col in X_train.columns if X_train[col].dtype == "object"]

# Preprocessing Numerical data
numerical_transformer = Pipeline(
    steps=[
        ("imputer", SimpleImputer(strategy="mean")),
    ]
)

# Preprocessing Categorical data
categorical_transformer = Pipeline(
    steps=[
        ("imputer", SimpleImputer(strategy="most_frequent")),
        ("onehot", OneHotEncoder(handle_unknown="ignore")),
    ]
)

column_transformer = ColumnTransformer(
    transformers=[
        ("num", numerical_transformer, num_cols),
        ("cat", categorical_transformer, cat_cols),
    ]
)

preprocess = Pipeline(steps=[("cts", column_transformer), ("scaler", StandardScaler())])

X_train = preprocess.fit_transform(X_train)
X_valid = preprocess.transform(X_valid)
X_test = preprocess.transform(X_test)

print("Shapes after Preprocessing.")
print("X_train", X_train.shape, "y_train", y_train.shape)
print("X_valid", X_valid.shape, " y_valid", y_valid.shape)
print("X_test ", X_test.shape)
```

### Solution to Imbalanced data issue: Cluster based undersampling

```python
from numpy.random import randint
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.cluster import KMeans, MiniBatchKMeans


class ClusterBasedUnderSampling(BaseEstimator, TransformerMixin):
    def fit(self, X, y):
        return self

    def transform(self, X, y):
        positive_X = X[y == 1]
        negative_X = X[y == 0]

        # initialize KMeans based on the size
        pos_size = len(positive_X)
        if pos_size > 1000:
            cluster = MiniBatchKMeans(n_clusters=pos_size)
        else:
            cluster = KMeans(n_clusters=pos_size)
        cluster.fit(negative_X)

        # use correct positives as positive train examples
        if ADDITIONAL_NEGATIVES in (0.5, 1, 2):
            additional_random_negatives = resample(
                negative_X,
                replace=False,
                n_samples=int(len(cluster.cluster_centers_) * ADDITIONAL_NEGATIVES),
                random_state=RANDOM_STATE,
            )

            X_negative = np.concatenate(
                [cluster.cluster_centers_, additional_random_negatives]
            )

        else:
            X_negative = cluster.cluster_centers_

        X_positive = positive_X
        # use clusters's centroids as negative train examples

        # concat positive and negatives to create X_train
        X_transformed = np.concatenate([X_positive, X_negative], axis=0)
        # create corresponding ones and zeros for y_train
        y_transformed = np.concatenate(
            [np.ones(len(X_positive)), np.zeros(len(X_negative))], axis=0
        )
        return X_transformed, y_transformed


undersampling = ClusterBasedUnderSampling()
undersampling.fit(X_train, y_train)
X_train, y_train = undersampling.transform(X_train, y_train)

print("Shapes after undersampling X_train based on cluster centroids.")
print("X_train", X_train.shape, "y_train", y_train.shape)
print("X_valid", X_valid.shape, "y_valid", y_valid.shape)
print("X_test ", X_test.shape)
```

# Modeling

```python
def print_and_return_results(model, X, y):
    # given a trained model, X, and y displays some
    # important metrics and returns them
    y_probs = model.predict_proba(X)
    y_preds = model.predict(X)

    # auc_score
    auc_score = roc_auc_score(y, y_probs[:, 1])

    # positive predicitions' ratio
    pos_count = (y_preds == 1).sum()
    pos_ratio = round(pos_count / len(y_valid), 3)

    # f1, precision, recall, and accuracy scores
    f_score = f1_score(y, y_preds)
    precision = precision_score(y, y_preds)
    recall = recall_score(y, y_preds)
    accuracy = accuracy_score(y, y_preds)

    print("AUC Score:", auc_score)
    print("Precision:", precision)
    print("Recall   :", recall)
    print("F1 Score :", f_score)
    print("Accuracy :", accuracy)
    print("Positive Predictions Ratio:", pos_ratio, "\n")

    return auc_score, f_score, accuracy, pos_ratio, precision, recall
```

```python
def train_model(classifier, grid_parameter):
    # use f1 score for scoring
    scorer = make_scorer(f1_score)

    # run grid search to find best params
    clf = GridSearchCV(classifier, grid_parameter, cv=3, scoring=scorer, n_jobs=-1)

    classifier_name = str(classifier).replace("()", "")
    title = f"{classifier_name} model"
    print(title)
    print(len(title) * "=")

    # Fit gridsearch model
    start = time.time()
    clf.fit(X_train, y_train)
    train_time = round((time.time() - start) / 60, 2)
    print(f"Fitting took {train_time} mins.")

    # obtain important metrics
    (
        auc_score,
        f_score,
        accuracy,
        pos_ratio,
        precision,
        recall,
    ) = print_and_return_results(clf, X_valid, y_valid)

    return [
        classifier_name,
        clf,
        train_time,
        f_score,
        auc_score,
        accuracy,
        pos_ratio,
        precision,
        recall,
    ]
```

```python
# consider the following classification methods with their corresponding parameter grids for optimization.
classifiers = {
    "log": [
        LogisticRegression(),
        [{"max_iter": [200, 370, 540], "penalty": ["l2", "None"]}],
    ],
    "xgb": [xgb.XGBClassifier(), [{"learning_rate": [0.001, 0.003, 0.01, 0.03, 0.1]}]],
    "rf": [
        RandomForestClassifier(),
        [
            {
                "max_depth": [11, 25, None],
                "n_estimators": [51, 101, 151, 201, 251, 301, 501, 701],
            }
        ],
    ],
    "nn": [
        MLPClassifier(),
        [
            {
                "hidden_layer_sizes": [[32, 32, 32], [32, 32], [64, 64]],
                "alpha": [0.0003, 0.03],
            }
        ],
    ],
    "knn": [KNeighborsClassifier(), [{"n_neighbors": [3, 5, 7]}]],
    "dt": [
        DecisionTreeClassifier(),
        [
            {
                "max_depth": [20, 50, 100, None],
                "criterion": ["gini", "entropy", "log_loss"],
            }
        ],
    ],
    "gnb": [GaussianNB(), [{"var_smoothing": [1e-9]}]],
    "svm": [SVC(probability=True), [{"gamma": ["auto", 0.025]}]],
    "ada": [
        AdaBoostClassifier(),
        [
            {
                "n_estimators": [101, 201, 301],
                "learning_rate": [0.003, 0.01, 0.03, 0.1],
            }
        ],
    ],
}
```

```python
# Train and compare all above models and save the results in df
df = pd.DataFrame(
    columns=[
        "classifier_class",
        "trained_model",
        "train_time",
        "F1_Score",
        "AUC",
        "Accuracy",
        "Pos Ratio",
        "Precision",
        "Recall",
    ]
)

for _, (classifier, grid_parameter) in classifiers.items():
    [
        classifier_class,
        trained_model,
        train_time,
        f_score,
        auc_score,
        accuracy,
        pos_ratio,
        precision,
        recall,
    ] = train_model(classifier, grid_parameter)

    df.loc[len(df)] = [
        classifier_class,
        trained_model,
        train_time,
        f_score,
        auc_score,
        accuracy,
        pos_ratio,
        precision,
        recall,
    ]
```

## Prepare Submission

```python
def prepare_submission(df, X_test):
    # Submission using Logistic Regression
    log_model = df.loc[
        df.classifier_class == "LogisticRegression"
    ].trained_model.values[0]

    # Predict probabilities
    clf_log_probs = log_model.predict_proba(X_test)

    # saves probabilities as csv
    np.savetxt(
        "submission/glmresults.csv",
        np.around(clf_log_probs[:, 1], decimals=4),
        delimiter=",",
    )

    # Get the best performing classifier, if the best performing
    # one is Logistic Regression, then obtain the second best
    ind = 0
    if df.iloc[0].classifier_class == "LogisticRegression":
        ind = 1
    best_model = df.iloc[ind].trained_model

    # Predict probabilities
    clf_best_probs = best_model.predict_proba(X_test)

    # saves probabilities as csv
    np.savetxt(
        "submission/nonglmresults.csv",
        np.around(clf_best_probs[:, 1], decimals=4),
        delimiter=",",
    )

    print("glmresults.csv and nonglmresults.csv are created in submission folder.")


prepare_submission(df, X_test)
```
